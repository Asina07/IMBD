##IMBD WEB SCRAPPING


```python
#import all the packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
#Url link of website
base_url='https://www.imdb.com/search/title/?release_date=2021-01-01,2021-12-31&sort=num_votes,desc&start='
film_name=[]
year_rele=[]
certificate=[]
time_duration=[]
genre=[]
rating=[]
score=[]
votes=[]
```

```python
for i in range(1,101,51):
  time.sleep(2)
  req=requests.get(base_url+str(i)+'&ref_=adv_nxt',headers={'Accept-Language':'en-US,en;q=0.5'})
  soup=BeautifulSoup(req.text,'html')
 
```

```python
# Obtain each movie in a list
my_movies=soup.find_all('div',class_='lister-item-content') 
```

```python
# Applying loop
for movie in my_movies:
  
# Name of movie
    mov_name=movie.find('h3',attrs={'class':'lister-item-header'}).a.text
    film_name.append(mov_name)

# Year of release
    year=movie.find('span',attrs={'class':'lister-item-year text-muted unbold'}).text
    year_rele.append(year)
 
#Certificate
    certi=movie.find('span',attrs={'class':'certificate'})
    if certi is None:
      certificate.append('No data')
    else:
      certificate.append(certi.text)

# Duration of film
    duration=movie.find('span',attrs={'class':'runtime'})
    if duration is None:
      time_duration.append('No data')
    else:
      time_duration.append(duration.text)
 
#Genre
    gen=movie.find('span',attrs={'class':'genre'}).text.strip()
    genre.append(gen)
 
# Rate
    rate=movie.find('div',attrs={'class':'inline-block ratings-imdb-rating'}).strong.text
    rating.append(ra)
 
# Metascore
    meta=movie.find('span',attrs={'class':'metascore'})
    if meta is None:
      meta_score.append('No data')
    else:
      meta_score.append(meta.text)
 
 
# Votes  
    vote=movie.find('span',attrs={'name':'nv'}).text
    votes.append(vote)  
 
```

```python
len(votes),len(meta),len(rating),len(genre),len(time_duration),len(certificate),len(year_rele),len(film_name)
```

```python
#Converting to DataFrame
data=pd.DataFrame({'name':film_name,'year':year_rele,'certifi':certificate,'time':time_duration,'Genre':genre,'Vote':votes,'Metascore':meta,'Rating':rating})
data
```

```python
data.to_csv('movie_data.csv',index=False)
```
